**1. ** Resize the lena or vibot image using nearest, bilinear and bicubic function.

You can define your function which accept the model and resize size from the user or just run a script.

```python
im_lena = plt.imread('images/lena-grey.bmp')
plt.imshow(im_lena, cmap=plt.cm.gray)
```

```python
#plt.figure()
#plt.imshow(im_resize, cmap=plt.cm.gray)
#plt.show()  

# Here we can choose the scaling factors for the nearset neighbour rescaling
scaleWidth = 2
scaleHeight = 1

# ___________________________________________#
# ----- Nearest neighbour interpolation ---- #
# ___________________________________________#

# We use Image from PIL so that we can get the image size
im_lena2 = Image.open('images/lena-grey.bmp')
lena_size = im_lena2.size

# We create an image of zeros which is scaled to the new image dimensions
newWidth = int(scaleWidth*lena_size[0])
newHeight = int(scaleHeight*lena_size[1])
newImage = np.zeros((newHeight, newWidth))

# We loop over the new image pixels and we use nearest neighbour interpolation

# We loop over the rows 
for i in range(0, newHeight-1):
    # We loop over the columns of each row 
    for j in range(0, newWidth-1):
        # We fill each pixel with the corresponding intensity of the nearest neighbour
        newImage[i][j] = im_lena[int(round(i/scaleHeight))][int(round(j/scaleWidth))]

# Displaying the image after resizing
plt.figure()
plt.title('Rescaled image using nearest neighbour method')
plt.imshow(newImage, cmap=plt.cm.gray)

# ___________________________________________#
# ----- Bilinear interpolation ---- #
# ___________________________________________#

bilinear = imresize(im_lena2, (300,600), interp ='bilinear', mode=None) 

plt.figure()
plt.title('Rescaled image using using bilinear interpolation')
plt.imshow(bilinear, cmap=plt.cm.gray)
# ___________________________________________#
# ----- Bicubic interpolation ---- #
# ___________________________________________#
bicubic = imresize(im_lena2, (300,600), interp = 'bicubic', mode = None)        

plt.figure()
plt.title('Rescaled image using using bicubic interpolation')
plt.imshow(bicubic, cmap=plt.cm.gray)
#print(bilImage)

```

**2. ** Similar to the function of `my_blending.m` that you saw before define your function of blending.

Blend the lena image and vibot image using simple, alpha and pyramid method.   

Pyramid blending method: 
1. Create a mask the same size as your input images 
2. Create Laplacian pyramid of the input images 
    2.1. Create Gaussian pyramids of the input images
    2.2. Create Laplacian pyramids using the Gaussian pyramids
3. Create Gaussian pyramid for the region mask
4. Blend the two laplacian pyramids using the pyramid of the mask as weigh 
    $$ BL = L_1 * (GM) + L_2 *(1 - GM)$$
5. Colapse the pyramid to get the final image 
    $$rescale(rescale(rescale(BL_3)+BL_2)+BL_1)+BL_0$$

Note : For pyramid blending approach, you can use:
    * transform.rescale for upscaling your images
    * transform.pyramid_gaussian for building your pyramids
    * The output of transform.pyramid_gaussian is a type generator and you need to convert it to a list :
    
        b = [a for a in Gaussian_pyramid]


```python
# Image Blending
# Simple image blending
orange = Image.open('images/orange.jpeg')
apple = Image.open('images/apple.jpeg')

orange_size = orange.size
apple_size = apple.size

orange = plt.imread('images/orange.jpeg')
apple = plt.imread('images/apple.jpeg')
output_image_simple = plt.imread('images/orange.jpeg')

# Implemenation of the simple blending method
for i in range(0, orange_size[0]-1):
    for j in range(0, orange_size[1]-1):
        if(j <= orange_size[0]/2):
            output_image_simple[i][j] = orange[i][j]
        else :
            output_image_simple[i][j] = apple[i][j]

# Alpha Blending 
apple_side = plt.imread('images/apple.jpeg')
orange_side = plt.imread('images/apple.jpeg')
alpha_output_image = plt.imread('images/apple.jpeg')

lower_side = 1
upper_side = 299
w = upper_side - lower_side
alpha = 1
beta = 0

# Implementation of the alpha blending method by choosing a specific window "w"
for i in range(0, orange_size[0]-1):
    for j in range(0, orange_size[1]-1):
        if(j < lower_side):
            orange_side[i][j] = apple[i][j]           
        if(j >= upper_side):
            orange_side[i][j] = 0  
        if(j >= lower_side and j < upper_side) :#
            orange_side[i][j] = alpha*apple[i][j] 
    alpha = alpha - 1/(w)
    
alpha = 1    
for i in range(0, orange_size[0]-1):
    for j in range(0, orange_size[1]-1):
        if(j < lower_side):
            apple_side[i][j] = 0    
        if(j >= upper_side):
            apple_side[i][j] = orange[i][j]
        if(j >= lower_side and j < upper_side) :
            apple_side[i][j] = (1-alpha)*orange[i][j] 
    alpha = alpha - 1/(w)  
    
for i in range(0, orange_size[0]-1):
    for j in range(0, orange_size[1]-1):
        alpha_output_image[i][j] = apple_side[i][j] + orange_side[i][j]
           
# Pyramid Blending
image_app = transform.rescale(apple, 256/300)
image_ora = transform.rescale(orange, 256/300)
mask = image_app

image_gauss1 = transform.pyramid_gaussian(image_app, 3)
gauss_pyramid = list(image_gauss1)
image_gauss2 = transform.pyramid_gaussian(image_ora, 3)
gauss_pyramid2 = list(image_gauss2)

# Laplace Pyramid of apple image
L3_a = gauss_pyramid[3] 
L2_a = gauss_pyramid[2] - transform.rescale(gauss_pyramid[3], 2)
L1_a = gauss_pyramid[1] - transform.rescale(gauss_pyramid[2], 2)
L0_a = gauss_pyramid[0] - transform.rescale(gauss_pyramid[1], 2)

# Laplace Pyramid of orange image
L3_o = gauss_pyramid2[3] 
L2_o = gauss_pyramid2[2] - transform.rescale(gauss_pyramid2[3], 2)
L1_o = gauss_pyramid2[1] - transform.rescale(gauss_pyramid2[2], 2)
L0_o = gauss_pyramid2[0] - transform.rescale(gauss_pyramid2[1], 2)

# Create the mask 
for i in range(0, 255):
    for j in range(0, 255):
        if(j <= 128):
            mask[i][j] = 0
        else:
            mask[i][j] = 1
            
# Laplace Pyramid for the mask
mask_gauss = transform.pyramid_gaussian(mask, 3)
gauss_pyramid3 = list(mask_gauss)
L3_m = gauss_pyramid3[3] 
L2_m = gauss_pyramid3[2] 
L1_m = gauss_pyramid3[1] 
L0_m = gauss_pyramid3[0] 

# Laplce Pyramid of the output image
L3_out = L3_a * L3_m + L3_o * (1 - L3_m)
L2_out = L2_a * L2_m + L2_o * (1 - L2_m)
L1_out = L1_a * L1_m + L1_o * (1 - L1_m)
L0_out = L0_a * L0_m + L0_o * (1 - L0_m)

# Output Image
out1 = transform.rescale(L3_out, 2)
out2 = transform.rescale(out1 + L2_out, 2)
out3 = transform.rescale(out2 + L1_out, 2)
out_image = L0_out + out3
```

Use the `orange.jpeg` and `apple.jpeg` imgages and blend them using simple, alpha and pyramid approach.
